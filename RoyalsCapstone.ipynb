# **Do racial disparities persist in cancer mortality rates among women over time?**

This project explores cancer mortality trends among Black and White women across multiple decades. By analyzing long-term patterns, the goal is to identify persistent racial disparities and understand how mortality rates have changed over time.

The findings highlight differences in health outcomes, emphasizing the need for equitable healthcare access, early detection, and targeted public health interventions.

# **Problem Statement:**

Despite overall improvements in cancer treatment and prevention, racial disparities in mortality rates remain significant. Black women consistently experience higher cancer mortality rates compared to White women across decades.

Understanding how these disparities have changed over time can help identify gaps in healthcare access, treatment effectiveness, and prevention efforts, ultimately informing more equitable health policies.

## **Data Definition:**

This dataset contains age-standardized cancer mortality rates for Black and White women across multiple decades. The data is structured to allow comparison over time and between racial groups.

Key variables include:
- Year or Decade
- Race (Black, White)
- Mortality Rate (per 100,000)

The dataset enables trend analysis, comparison of mortality rates, and calculation of disparities between groups.

```python
# Clean column names
ga_cancer.columns = ga_cancer.columns.str.strip().str.lower().str.replace(" ", "_")

# Check missing values
ga_cancer.isnull().sum()

# Drop missing rows (if needed)
ga_cancer = ga_cancer.dropna()

# Preview cleaned data
ga_cancer.head(10)
```

```python
from scipy.stats import chi2_contingency

# Group deaths by time period
table = ga_cancer.groupby("years")[[
    "blackfemale_deaths_allsites",
    "whitefemale_deaths_allsites"
]].sum()

# Run chi-square test
chi2, p, dof, expected = chi2_contingency(table)

print("p-value:", p)
```

```python
# Create simple time order
ga_cancer["period_order"] = ga_cancer["years"].replace({
    ga_cancer["years"].unique()[0]: 1,
    ga_cancer["years"].unique()[1]: 2
})

from sklearn.linear_model import LinearRegression

X = ga_cancer[["period_order"]]
y_black = ga_cancer["blackfemale_rate_allsites"]

model = LinearRegression()
model.fit(X, y_black)

print("Trend (Black women):", model.coef_[0])
```
